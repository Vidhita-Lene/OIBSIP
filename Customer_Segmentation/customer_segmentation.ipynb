### 4 Data Cleaning

## 5 Feature Engineering

#### Total Spending

```python
data['Total_Spending'] = (
    data['MntWines']
    + data['MntFruits']
    + data['MntMeatProducts']
    + data['MntFishProducts']
    + data['MntSweetProducts']
    + data['MntGoldProds']
)
```

#### Total Purchase

```python
data['Total_Purchases'] = (
    data['NumWebPurchases']
    + data['NumCatalogPurchases']
    + data['NumStorePurchases']
)
```

## 8 Select Features for Clustering

```python
features = data[['Income','Total_Spending','Total_Purchases']]
features
```

## 9 Feature Scaling

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)
scaled_data
```

## 10 Determine Optimal Clusters (Elbow Method)

```python
inertia = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
plt.figure(figsize=(6,4))
plt.plot(range(1,10), inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

## 11 K-Means Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
data.head()
```

## 12 Visualizing Customer Segments

#### Income vs Spending by Cluster

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=data['Income'],
    y=data['Total_Spending'],
    hue=data['Cluster'],
    palette='Set1'
)
plt.title("Customer Segmentation")
plt.show()
```

#### Spending vs Purchases

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    x=data['Total_Purchases'],
    y=data['Total_Spending'],
    hue=data['Cluster'],
    palette='Set2'
)

plt.title("Customer Purchase Behavior")
plt.show()
```

#### Number of customers in each cluster

```python
sns.countplot(x=data['Cluster'])

plt.title("Number of Customers in Each Cluster")
plt.show()
```
